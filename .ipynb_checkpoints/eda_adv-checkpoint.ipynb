**Parse Data, Source :  https://www.kaggle.com/datasets/olistbr/brazilian-ecommerce**

````python
# Parse Data, Source :  https://www.kaggle.com/datasets/olistbr/brazilian-ecommerce

print('Locating Datasets....')
try:
    data_dir = kagglehub.dataset_download('olistbr/brazilian-ecommerce')
    print(f'Dataset Located at: {data_dir}\nGood to goo...')
except Exception as e: print(f'An error eccured. Details: {e}')
````

**load all datasets**

**olist_orders_datasets.csv :**

````python
# olist_orders_datasets.csv
print('olist_orders_datasets.csv: ')
df_orders = pd.read_csv(os.path.join(data_dir, 'olist_orders_dataset.csv'))
df_orders.head()
````

**olist_customers_datasets.csv :**

````python
# olist_customers_datasets.csv
print('olist_customers_datasets.csv: ')
df_customers = pd.read_csv(os.path.join(data_dir, 'olist_customers_dataset.csv'))
df_customers.head()
````

**olist_order_items.csv :**

````python
# olist_order_items.csv
print('olist_order_items.csv: ')
df_order_items = pd.read_csv(os.path.join(data_dir,'olist_order_items_dataset.csv'))
df_order_items.head()
````

**olist_product_dataset :**

````python
# olist_product_dataset
print('olist_product_dataset.csv: ')
df_products = pd.read_csv(os.path.join(data_dir,'olist_products_dataset.csv'))
df_products.head()
````

**olsit_seller_dataset.csv**

````python
# olsit_seller_dataset.csv
print('olist_seller_dataset.csv: ')
df_sellers = pd.read_csv(os.path.join(data_dir,'olist_sellers_dataset.csv'))
df_sellers.head()
````

````python
# olist_order_payments_datasets.csv
print('olist_order_payments_datasets.csv')
df_order_payments = pd.read_csv(os.path.join(data_dir,'olist_order_payments_dataset.csv'))
df_order_payments.head()

````

**olist_review_dataset.csv**

````python
# olist review dataset.csv
print('olist review dataset.csv: ')
df_order_reviews = pd.read_csv(os.path.join(data_dir,'olist_order_reviews_dataset.csv'))
df_order_reviews.head()
````

**olist_category_translation_dataset.csv :**

````python
# olist_category_translation_dataset.csv
print('olist_category_translation.csv: ')
df_category_translation = pd.read_csv(os.path.join(data_dir,'product_category_name_translation.csv'))
df_category_translation.head()
````

**merge dataframes into a single master dataframe**

````python
# start with order and customers
df = pd.merge(df_orders, df_customers, on='customer_id', how='left')

# add order items and seller info
df = pd.merge(df, df_order_items, on='order_id', how='left')

# add product info
df = pd.merge(df, df_products, on='product_id', how='left')

# add payment info
df = pd.merge(df, df_order_payments, on='order_id', how='left')

# add review info
df = pd.merge(df, df_order_reviews, on='order_id', how='left')

# add seller info
df = pd.merge(df, df_sellers, on='seller_id', how='left')

# add category translation
df = pd.merge(df, df_category_translation, on='product_category_name', how='left')

print(f'Shape of Merge Dataframes: {df.shape}')
df.head()
````

### 1.3.2. Data Cleaning

**Missing Value**

jangan anggap nilai kosong adalah masalah yang bikin bingung, tapi tanyakan dulu alasan dibalik ke kosongan itu.

*Inti dari handling missing values bukan cuman biar ga error, tapi untuk jaga validitas data*

````python
total_null = df.isnull().sum(); percent_null = (df.isnull().sum() / df.shape[0]) * 100
df_null = pd.DataFrame({
    'Total Null': total_null,
    'Persentase Null (%)': percent_null
}).round(2); df_null.sort_values(by='Persentase Null (%)', ascending=False)
````

````python
df_null.sort_values(by='Total Null', ascending=True)['Total Null'].plot(
    kind='barh',
    color='skyblue',
    figsize=(10, 8)
)

plt.title('Jumlah Nilai Null per Kolom')
plt.xlabel('Jumlah Null')
plt.ylabel('Kolom')
plt.tight_layout()
plt.show()
````

**Missing Values != Bias**

missing reviews, kaya `review_comment_title` dan `review_comment_massage` banyak bet kosongnya, tapi ini *wajar*. karena emang kebanyakan orang kasih bintang doang, ga nulis komentas. yaa jangan dibuang,itu gaakan bikin analisis kita ngawur, karena emang seharusnya itu kosong.

terus juga coba kita liat kolom `order_delivered_customer_date`. banyak juga null nya, tapi ini juga wajar, karena kebayakan status ordernya cencelled, shopped/processing atau belum nyampe. jadi wajar aja tanggal nya belum ada. ini coba kita buktikan 

````python
# Investigate missing delivery dates
missing_delivery_date_df = df[df['order_delivered_customer_date'].isnull()]
print(f'missing value in delivary data : {missing_delivery_date_df.shape[0]}')

status_counts = missing_delivery_date_df['order_status'].value_counts()
print("Order statuses for orders with missing delivery dates:")
print(status_counts)
````

see?. null karena emang status ordernya belum nyampe atau cencelled. dia bakal bias atau bisa diapus ketika delivered atau udah nyampe, tapi tanggal masih ga kecatat.

````python
# Buat kondisi untuk baris yang mau dihapus
mask = (df['order_status'] == 'delivered') & (df['order_delivered_customer_date'].isnull())

# Hapus baris tersebut dari df
df = df[~mask]
````

````python
print(status_counts)
````

all done for now, kita udah apus yang kira kira perlu ajaa.

**Data Types**

coba kita cek tipe data tiap kolom

````python
df.info()
````

ada beberapa permasalah disini, dan sering juga jadi masalah, yaitu datetime

````python
# Convert timestamp columns to datetime objects
timestamp_cols = ['order_purchase_timestamp', 'order_approved_at', 
                  'order_delivered_carrier_date', 'order_delivered_customer_date', 
                  'order_estimated_delivery_date', 'shipping_limit_date',
                  'review_creation_date', 'review_answer_timestamp']

for col in timestamp_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')

# Ensure numeric columns are of the correct type
numeric_cols = ['price', 'freight_value']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df.info()
````

### 2.1. Temporal Anlaysis

**Growth Trends**

kita mau tau penjualan per bulannya, jadi kita tinggal gabungin semua harga (`price`) berdasarkan bulan pembelian

````python
# set kolom waktu sebagai index
df_ts = df.set_index('order_purchase_timestamp')

# hitung total revenue per bulan
monthly_revenue = df_ts['price'].resample('M').sum()

# visualisasi
plt.figure(figsize=(15,7)); monthly_revenue.plot(kind='line', marker='o', color='b')
plt.title('Olist Monthly Revenue', fontsize=15)
plt.xlabel('Month', fontsize=12); plt.ylabel('Total Revenue', fontsize=12)
plt.grid(True); plt.show()
````

okey grafiknya agak anek, gak mungkin tiba tiba nol gitu. kemungkinan besar, bulan itu masih awal awal dan belum terinput datanya, so kita bakal ambil sampe agustus aja.

````python
monthly_revenue = monthly_revenue[monthly_revenue.index <= '2018-08-31']

# visualisasi
plt.figure(figsize=(15,7)); monthly_revenue.plot(kind='line', marker='o', color='b')
plt.title('Olist Monthly Revenue', fontsize=15)
plt.xlabel('Month', fontsize=12); plt.ylabel('Total Revenue', fontsize=12)
plt.grid(True); plt.show()
````

dari grafik revenue kita dapatkan bahwa tren penjualan itu condong naik dari waktu ke waktu, menandakan bisnis mereka terus berkembang maju. 

namun pertanyaannya apakah itu naik karena olist terus berkembang, atau memang tren e com atau ekon disana emang lagi berkembang aja, kalo emang iya yaa, emang wajar, bukan karena mereka karena market.

### 3.1. Order Lifecycle – From Buy to Arrived


beberapa kolom penting disini:

`order_approved_at` : pembayaran disetujui

`order_purchase_timestamp` : after payment

`order_delivered_carrier_date` : pengemasan oleh seller (dikirim ke seller)

`order_delivered_customer_date` : barang diterima



Pertanyaan:

**Time Approval** atau waktu dari orang beli sampe disetujui. ini berfokus ke ngukur efisiensi sistem pembayaran olist.

rumus: setelah pembayaran s/d di acc. 

`time_to_approval` = `order_approved_at` - `order_purchase_timestamp`

**Seller Processing Time**, dikemas. yang diukur performa seller.

rumus : setelah di acc s/d pengemasan selesai.

`seller_processing_time` = `order_delivered_carrier_date` - `order_approved_at`

**Carrier Shipping Time**, waktu pengiriman, mengukur performa ekspedisi atau logistik

rumus: setelah diterima logistik s/d pesanan diterima

`carrier_shipping_time` = `order_delivered_carrier_data` - `order_delivered_customer_date`

**Total Fulfillment Time**, waktu abis bayar sampe pesanan diterima

rumus: setelah bayar s/d diterima customer

`total_fulfillment_time` = `order_purchase_timestamp` - `order_delivered_customer_date`

cara analisis nya kita cuma perlu ngitung waktu kaya gini:

```python

# (Tgl B - Tgl A).dt.total_seconds() / 86400  --> hasilnya dalam hari

```

tapi...

anlisis ini valid jika status order delivered, kalo masih shipped/cencelled dll, waktunya gaakan valid. dan emang rata rata masi null.

````python
# Filter hanya order yang berhasil dikirim
delivered_df = df[df['order_status'] == 'delivered'].dropna(subset=timestamp_cols)
````

````python
# Hitung masing-masing waktu
delivered_df['time_to_approval'] = (delivered_df['order_approved_at'] - delivered_df['order_purchase_timestamp']).dt.total_seconds() / 86400
delivered_df['seller_processing_time'] = (delivered_df['order_delivered_carrier_date'] - delivered_df['order_approved_at']).dt.total_seconds() / 86400
delivered_df['carrier_shipping_time'] = (delivered_df['order_delivered_customer_date'] - delivered_df['order_delivered_carrier_date']).dt.total_seconds() / 86400
delivered_df['total_fulfillment_time'] = (delivered_df['order_delivered_customer_date'] - delivered_df['order_purchase_timestamp']).dt.total_seconds() / 86400
````

### 3.3. What comes after this Delay?

pertanyaan: 7 persen keterlambatan, impact nya besar ga ke kepuasan pelanggan?

jawaban: analisa hubungan `delivery_delay` dan `review_score`.

caranya:

buat fitur `is_late`(terlambat atau engga), 

lakukan **t-test** -> bandingin rata rata rating antara pesanan terlambat dan tepat waktu.

**insight :**

| Aspek                              | Temuan Utama                                       | Penjelasan Singkat                                                                 | Insight Tindakan                                                              |
|------------------------------------|----------------------------------------------------|-------------------------------------------------------------------------------------|--------------------------------------------------------------------------------|
| Delay Ringan & Review Tinggi       | Skor 4–5 tetap dominan saat delay < 10 hari        | Pelanggan masih memaklumi keterlambatan kecil, mungkin karena faktor lain positif  | Fokus utama bukan hanya kecepatan, tapi pengalaman menyeluruh                |
| Delay Berat Turunkan Review        | Delay > 20 hari banyak dapat skor 1–2              | Keterlambatan signifikan langsung berdampak pada kepuasan pelanggan                | Prioritaskan perbaikan untuk pesanan dengan delay ekstrem                     |
| Datang Lebih Awal ≠ Review Tinggi | Ada review 1–2 meski barang datang lebih cepat     | Kualitas produk/seller lebih penting daripada kecepatan pengiriman                 | Tingkatkan kualitas produk dan pelayanan seller                               |
| Review Score Diskrit               | Terdistribusi dalam garis horizontal 1–5           | Karena review score berbentuk bilangan bulat                                       | Visualisasi diskrit tetap valid dan mudah dibaca                              |

---

**kesimpulan :**

Sebagian besar pelanggan tetap puas meski ada sedikit keterlambatan. Namun, jika keterlambatan melebihi 20 hari, risiko ketidakpuasan meningkat tajam. Sementara itu, pengiriman lebih cepat tidak otomatis meningkatkan review — kualitas produk dan layanan tetap krusial.

````python
from scipy.stats import ttest_ind

delivered_df['is_late'] = delivered_df['delivered_delay'] > 0

late_reviews = delivered_df[delivered_df['is_late'] == True]['review_score']
on_time_reviews = delivered_df[delivered_df['is_late'] == False]['review_score']

t_stat, p_value = ttest_ind(late_reviews, on_time_reviews, equal_var=False)  # Gunakan Welch’s t-test

print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")
````

T-statistic: -92.23 (nilai absolut sangat besar)

P-value: 0.0000 (lebih kecil dari 0.05 → signifikan)

Nilai p-value yang sangat kecil mengindikasikan bahwa perbedaan antara dua kelompok bukan karena kebetulan.

Dengan kata lain, keterlambatan pengiriman memang menyebabkan turunnya review score secara signifikan.

````python
plt.figure(figsize=(10,6))
sns.boxplot(x='is_late', y='review_score', data=delivered_df, palette='Set2')
plt.title('Review Score by Delivery Timeliness', fontsize=16)
plt.xlabel('Is Late (True = Terlambat)', fontsize=12)
plt.ylabel('Review Score', fontsize=12)
plt.grid(True)
plt.show()
````

### 4.1. RFM Segmentation: Transaction defines Behavior

**RFM** (Recency Frequency Monetary) segmentation bertujuan untuk klasifikasi pelanggan berdasarkan perilaku transaksi mereka

**ini bisa mendefinisikan :**

pelanggan bernilai tinggi (Champion, Loyal Customer)

pelanggan pertumbuhan (Potential Loyalist, New Customer)

Resiko Churn (At Risk, Hibernating, Lost).

**Langkah :**

1. Recency: ngukur bearapa hari sejak pembelian terakhir seorang pelanggan hingga snapshot date (tanggal acuan tertentu / treshold). semakin tinggi nilainya, semakin baru pembeliannya, semakin bagus skornya.

2. Frequecy: ngukur berapa sering atau berapa kali lakukan transaksi. semakin sering beli, semakin tinggi nilainya.

3. Monetary: ngukur berapa total uang dikeluarkan pelanggan. sebara banyak udah belanjain duitnya

4. scoring & segmentation. setiap metrik diberi skor 1-4 berdasarkan kuartil skor digabungkan menjadi format RFM string (cth. 443). kemudian di klasifikasin ke dalam 8 segment utama menggunakan 


---

````python
# --- RFM Calculation ---

# Set a snapshot date for recency calculation (e.g., one day after the last order)
snapshot_date = df['order_purchase_timestamp'].max() + pd.Timedelta(days=1)

# Aggregate data at the customer level (using customer_unique_id)
rfm_data = df.groupby('customer_unique_id').agg({
    'order_purchase_timestamp': lambda x: (snapshot_date - x.max()).days,
    'order_id': 'nunique',
    'price': 'sum'
})

# Rename columns for clarity
rfm_data.rename(columns={'order_purchase_timestamp': 'Recency',
                         'order_id': 'Frequency',
                         'price': 'Monetary'}, inplace=True)

# --- RFM Scoring ---

# Create labels and bins for scoring
r_labels = range(4, 0, -1) # Higher recency (lower days) gets a higher score
f_labels = range(1, 5)
m_labels = range(1, 5)

# Assign scores based on quantiles
rfm_data = pd.qcut(rfm_data, q=4, labels=r_labels, duplicates='drop')
rfm_data = pd.qcut(rfm_data['Frequency'].rank(method='first'), q=4, labels=f_labels, duplicates='drop')
rfm_data = pd.qcut(rfm_data['Monetary'], q=4, labels=m_labels, duplicates='drop')

# --- RFM Segmentation ---

# Combine scores to create an RFM segment string
def join_rfm(x): return str(x) + str(x) + str(x)
rfm_data = rfm_data.apply(join_rfm, axis=1)

# Calculate a total RFM score
rfm_data = rfm_data].sum(axis=1)

# Define segments based on the scores
# This mapping can be customized based on business logic
segment_map = {
    r'[3-4][3-4][3-4]': 'Champions',
    r'[2-4][3-4][1-4]': 'Loyal Customers',
    r'[3-4][1-2][3-4]': 'Potential Loyalists',
    r'[3-4][1-2][1-2]': 'New Customers',
    r'[2-3][1-2][1-4]': 'Promising',
    r'[1-2][3-4][1-4]': 'At Risk',
    r'[1-2][1-2][3-4]': 'Hibernating',
    r'[1-2][1-2][1-2]': 'Lost'
}

rfm_data = rfm_data.replace(segment_map, regex=True)

# Display the resulting RFM table
display(rfm_data.head())

# Visualize the distribution of segments
segment_counts = rfm_data.value_counts().sort_values(ascending=True)
plt.figure(figsize=(12, 8))
segment_counts.plot(kind='barh', color=sns.color_palette('mako', len(segment_counts)))
plt.title('Customer Segment Distribution', fontsize=16)
plt.xlabel('Number of Customers', fontsize=12)
plt.ylabel('Segment', fontsize=12)
plt.show()
````
