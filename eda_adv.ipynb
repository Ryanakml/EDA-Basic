**merge dataframes into a single master dataframe**

````python
# start with order and customers
df = pd.merge(df_orders, df_customers, on='customer_id', how='left')

# add order items and seller info
df = pd.merge(df, df_order_items, on='order_id', how='left')

# add product info
df = pd.merge(df, df_products, on='product_id', how='left')

# add payment info
df = pd.merge(df, df_order_payments, on='order_id', how='left')

# add review info
df = pd.merge(df, df_order_reviews, on='order_id', how='left')

# add seller info
df = pd.merge(df, df_sellers, on='seller_id', how='left')

# add category translation
df = pd.merge(df, df_category_translation, on='product_category_name', how='left')

print(f'Shape of Merge Dataframes: {df.shape}')
df.head()
````

### 1.3.2. Data Cleaning

**Missing Value**

jangan anggap nilai kosong adalah masalah yang bikin bingung, tapi tanyakan dulu alasan dibalik ke kosongan itu.

*Inti dari handling missing values bukan cuman biar ga error, tapi untuk jaga validitas data*

````python
total_null = df.isnull().sum(); percent_null = (df.isnull().sum() / df.shape[0]) * 100
df_null = pd.DataFrame({
    'Total Null': total_null,
    'Persentase Null (%)': percent_null
}).round(2); df_null.sort_values(by='Persentase Null (%)', ascending=False)
````

````python
df_null.sort_values(by='Total Null', ascending=True)['Total Null'].plot(
    kind='barh',
    color='skyblue',
    figsize=(10, 8)
)

plt.title('Jumlah Nilai Null per Kolom')
plt.xlabel('Jumlah Null')
plt.ylabel('Kolom')
plt.tight_layout()
plt.show()
````

**Missing Values != Bias**

missing reviews, kaya `review_comment_title` dan `review_comment_massage` banyak bet kosongnya, tapi ini *wajar*. karena emang kebanyakan orang kasih bintang doang, ga nulis komentas. yaa jangan dibuang,itu gaakan bikin analisis kita ngawur, karena emang seharusnya itu kosong.

terus juga coba kita liat kolom `order_delivered_customer_date`. banyak juga null nya, tapi ini juga wajar, karena kebayakan status ordernya cencelled, shopped/processing atau belum nyampe. jadi wajar aja tanggal nya belum ada. ini coba kita buktikan 

````python
# Investigate missing delivery dates
missing_delivery_date_df = df[df['order_delivered_customer_date'].isnull()]
print(f'missing value in delivary data : {missing_delivery_date_df.shape[0]}')

status_counts = missing_delivery_date_df['order_status'].value_counts()
print("Order statuses for orders with missing delivery dates:")
print(status_counts)
````

see?. null karena emang status ordernya belum nyampe atau cencelled. dia bakal bias atau bisa diapus ketika delivered atau udah nyampe, tapi tanggal masih ga kecatat.

````python
# Buat kondisi untuk baris yang mau dihapus
mask = (df['order_status'] == 'delivered') & (df['order_delivered_customer_date'].isnull())

# Hapus baris tersebut dari df
df = df[~mask]
````

### 3.1. Order Lifecycle ‚Äì From Buy to Arrived


beberapa kolom penting disini:

`order_purchase_timestamp` : after payment

`order_delivered_carrier_date` : pengemasan oleh seller (dikirim ke seller)

`order_delivered_customer_date` : barang diterima



Pertanyaan:

**Time Approval** atau waktu dari orang beli sampe disetujui. ini berfokus ke ngukur efisiensi sistem pembayaran olist.

rumus: setelah pembayaran s/d di acc. 

`time_to_approval` = `order_approved_at` - `order_purchase_timestamp`

**Seller Processing Time**, dikemas. yang diukur performa seller.

rumus : setelah di acc s/d pengemasan selesai.

`seller_processing_time` = `order_delivered_carrier_date` - `order_approved_at`

**Carrier Shipping Time**, waktu pengiriman, mengukur performa ekspedisi atau logistik

rumus: setelah diterima logistik s/d pesanan diterima

`carrier_shipping_time` = `order_delivered_carrier_data` - `order_delivered_customer_date`

**Total Fulfillment Time**, waktu abis bayar sampe pesanan diterima

rumus: setelah bayar s/d diterima customer

`total_fulfillment_time` = `order_purchase_timestamp` - `order_delivered_customer_date`

**kesimpulan :**

Sebagian besar pelanggan tetap puas meski ada sedikit keterlambatan. Namun, jika keterlambatan melebihi 20 hari, risiko ketidakpuasan meningkat tajam. Sementara itu, pengiriman lebih cepat tidak otomatis meningkatkan review ‚Äî kualitas produk dan layanan tetap krusial.

````python
from scipy.stats import ttest_ind

delivered_df['is_late'] = delivered_df['delivered_delay'] > 0

late_reviews = delivered_df[delivered_df['is_late'] == True]['review_score']
on_time_reviews = delivered_df[delivered_df['is_late'] == False]['review_score']

t_stat, p_value = ttest_ind(late_reviews, on_time_reviews, equal_var=False)  # Gunakan Welch‚Äôs t-test

print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")
````

T-statistic: -92.23 (nilai absolut sangat besar)

P-value: 0.0000 (lebih kecil dari 0.05 ‚Üí signifikan)

Nilai p-value yang sangat kecil mengindikasikan bahwa perbedaan antara dua kelompok bukan karena kebetulan.

### 4.1. RFM Segmentation: Transaction defines Behavior

**RFM** (Recency Frequency Monetary) segmentation bertujuan untuk klasifikasi pelanggan berdasarkan perilaku transaksi mereka

**ini bisa mendefinisikan :**

pelanggan bernilai tinggi (Champion, Loyal Customer)

pelanggan pertumbuhan (Potential Loyalist, New Customer)

Resiko Churn (At Risk, Hibernating, Lost).

**Langkah :**

1. Recency: ngukur bearapa hari sejak pembelian terakhir seorang pelanggan hingga snapshot date (tanggal acuan tertentu / treshold). semakin tinggi nilainya, semakin baru pembeliannya, semakin bagus skornya.

2. Frequecy: ngukur berapa sering atau berapa kali lakukan transaksi. semakin sering beli, semakin tinggi nilainya.

3. Monetary: ngukur berapa total uang dikeluarkan pelanggan. sebara banyak udah belanjain duitnya

4. scoring & segmentation. setiap metrik diberi skor 1-4 berdasarkan kuartil skor digabungkan menjadi format RFM string (cth. 443). kemudian di klasifikasin ke dalam 8 segment utama menggunakan 

````python
# 1. Hitung RFM Metrics (Recency, Frequency, Monetary)

# Tentukan snapshot date (1 hari setelah order terakhir)
snapshot_date = df['order_purchase_timestamp'].max() + pd.Timedelta(days=1)

# Hitung Recency, Frequency, dan Monetary per customer
rfm = df.groupby('customer_unique_id').agg({
    'order_purchase_timestamp': lambda x: (snapshot_date - x.max()).days,
    'order_id': 'nunique',
    'price': 'sum'
}).reset_index()

# Rename kolom
rfm.columns = ['customer_unique_id', 'Recency', 'Frequency', 'Monetary']
````

````python
# 2. Skoring RFM Berdasarkan Kuartil

# Skor Recency: semakin kecil recency, semakin baik (score tinggi)
rfm['R_score'] = pd.qcut(rfm['Recency'], 4, labels=[4, 3, 2, 1])

# Skor Frequency dan Monetary: semakin besar nilainya, semakin baik
rfm['F_score'] = pd.qcut(rfm['Frequency'].rank(method='first'), 4, labels=[1, 2, 3, 4])
rfm['M_score'] = pd.qcut(rfm['Monetary'], 4, labels=[1, 2, 3, 4])
````

````python
# 3. Buat RFM Segment dan Total Score

# Gabungkan menjadi RFM segment
rfm['RFM_Segment'] = rfm['R_score'].astype(str) + rfm['F_score'].astype(str) + rfm['M_score'].astype(str)

# Hitung total RFM Score
rfm['RFM_Score'] = rfm[['R_score', 'F_score', 'M_score']].astype(int).sum(axis=1)
````

````python
# 4. Mapping ke Kategori Segmentasi

# Buat segment map berdasarkan pola regex
segment_map = {
    r'[3-4][3-4][3-4]': 'Champions',
    r'[2-4][3-4][1-4]': 'Loyal Customers',
    r'[3-4][1-2][3-4]': 'Potential Loyalists',
    r'[3-4][1-2][1-2]': 'New Customers',
    r'[2-3][1-2][1-4]': 'Promising',
    r'[1-2][3-4][1-4]': 'At Risk',
    r'[1-2][1-2][3-4]': 'Hibernating',
    r'[1-2][1-2][1-2]': 'Lost'
}

# Klasifikasi segmen
rfm['Segment'] = rfm['RFM_Segment'].replace(segment_map, regex=True)
````
